```python
import allel
import numpy as np
import pandas as pd
```

## JPT population 


#### JPT data with allel package ###

```python
JPTdata = allel.read_vcf('JPT_SNPs_only.vcf', fields='*') ## Read the VCF file using allel 

JPTgenotypes=JPTdata['calldata/GT'] ## extract the genotypes array 

JPTarray=allel.GenotypeArray(JPTgenotypes) 
JPTarray## convert the genotypes data frame into a genotype array 

JPTgenotypes.n_samples ## count the total number of samples that will be used later to calculate genotypes frequencies 
```

#### JPT data extracted with vcf tools ###

## vcf tools command 

```python
##  "--vcf JPT_SNPs.vcf --hardy --out JPT.hwe "
```

```python
jpt_data_frame = pd.read_csv ('jpt.hwe',delimiter="\t") ## Read the hwe file extrcated from vcf tools 
jpt_data_frame
```

```python
keep_col = ['OBS(HOM1/HET/HOM2)'] ## extract only the columns with the genotypes 
jpt_data_frame=jpt_data_frame[keep_col] ## save it as a dataframe 
jpt_data_frame
```

```python
JPT=jpt_data_frame['OBS(HOM1/HET/HOM2)'].str.split('/', expand=True)## split the data frame to get seperate column for each genotype 
```

```python
JPT=JPT.rename(columns={JPT.columns[0]:"JPT_0|0 "})## rename the columns with the country code and the corresponding genotype symbols 
JPT=JPT.rename(columns={JPT.columns[1]:"JPT_0|1 "})
JPT=JPT.rename(columns={JPT.columns[2]:"JPT_1|1 "})
```

```python
JPT
```

```python
JPT = JPT.astype(int) #convert the date frame date into int 
JPT.div(104).## divide the genotypes by the number of samples (extracted earlier with allel ) to get the genotype frequencies 
```
